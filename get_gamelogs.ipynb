```python
import requests
import zipfile
import io
import pandas as pd
import numpy as np
```

```python
url = 'https://www.retrosheet.org/gamelogs/gl2010_18.zip'

r = requests.get(url, stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
ext = z.extractall()
```

```python
glhead = pd.read_csv('GLHEADER.CSV',header=None)
# !cp GLHEADER.TXT GLHEADER.CSV
```

```python
gl = pd.DataFrame([])
for n in range(2010,2019):
    year = str(n)
    filetxt = 'GL'+year+'.TXT'
    filecsv = 'GL'+year+'.CSV'
    !cp {filetxt} {filecsv}
    glyr = pd.read_csv(filecsv,header=0,names=list(glhead[0]))
    gl = gl.append(glyr)
    
gl.head()

# CAUTION make sure .CSV GL and HEADER files have been created before removing below    
!rm GL*.TXT
!rm GLHEADER.TXT
```

```python
gl['date'] = pd.to_datetime(gl['date'],format='%Y%m%d')
gl.head()
```
